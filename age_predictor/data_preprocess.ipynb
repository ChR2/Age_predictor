```python
# read the ignores, get rid of new lines
ignored_list = ignored_list.split('\n')

```

```python
ignores = [x.replace('.jpg', '') for x in ignored_list]
```

```python
def rm_ignores(folder):
    ignored = 0
    for f in os.listdir(folder):
        if f.split('.')[0] in ignores:
            ignored = ignored + 1
            os.rename(folder+'/'+f, '../ignored/'+f)
            # print(f)
    print('Ignoring {i} in {f}'.format(i=ignored, f=folder))

rm_ignores('test')
rm_ignores('train')
rm_ignores('valid')
```

```python
def bin_images(folder):
    fname = 'gt_avg_{}.csv'.format(folder)
    df = pd.read_csv(fname)
    df.file_name = df.file_name.apply(lambda x: x.split('.')[0])


    for f in os.listdir(folder):
        if 'face' in f:
            key = f.split('.')[0]
            if key in list(df['file_name']):
                age = str(int(df[df['file_name'] == key]['real_age']))
                print(f, age)
                directory = 'bins/'+folder+'/'+age+'/'
                if not os.path.exists(directory):
                    os.makedirs(directory)
                shutil.copy2(folder+'/'+f, directory)
                
bin_images('test')
bin_images('train')
bin_images('valid')
```
